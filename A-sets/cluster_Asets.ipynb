```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
s1 = pd.read_csv('a1.csv')
s11 = s1.dropna(axis = 0, how='all')
s11.head()
```

```python
x1 = s11.iloc[:,[0,1]]
x1.head()
```

```python
def plot():
    for a in range(15):
        plt.scatter(x1['Unnamed: 0'],x1['Unnamed: 1'], s = 1, cmap='viridis', color = 'black')
    
plot()
plt.title('Data')
plt.legend()
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x1,method = 'ward'))
plt.title('dendrogram')
plt.xlabel('data')
plt.ylabel('eucledian distance')
#plt.update_layout(width=800, height=500)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 20, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(x1)
y_hc
```

```python
x1=np.array(x1)
def plot():
    for a in range(20):
        plt.scatter(x1[y_hc == a, 0], x1[y_hc == a, 1], s = 10, cmap='rainbow')
        
plot()
plt.title('Clusters')
plt.legend()
plt.show()
```

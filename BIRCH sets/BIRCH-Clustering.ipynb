```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import Birch
```

```python
d = pd.read_csv('birch1.csv')
d.head()
```

```python
d.describe()
```

```python
d.isnull().sum()
```

```python
d1 = d.dropna(axis = 0, how='any')
d1.isnull().sum()
```

```python
plt.scatter(d1['x'], d1['y'], s=.01, color = 'black')
```

```python
brc = Birch(n_clusters=100, branching_factor=50,threshold=6500)
```

```python
brc.fit(d1)
```

```python
labels = brc.predict(d1)
```

```python
plt.scatter(d1.iloc[:,0], d1.iloc[:,1], c=labels, cmap='rainbow', alpha=0.7, s=.01)
```
